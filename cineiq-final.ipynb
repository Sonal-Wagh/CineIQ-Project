```python
MIN_RATINGS = 20
mc = ratings['movieId'].value_counts()
uc = ratings['userId'].value_counts()

ratings_filtered = ratings[
    ratings['movieId'].isin(mc[mc >= MIN_RATINGS].index) &
    ratings['userId'].isin(uc[uc >= MIN_RATINGS].index)
].copy()

print(f"Before : {len(ratings):,}")
print(f"After  : {len(ratings_filtered):,}")
print(f"Users  : {ratings_filtered['userId'].nunique():,}")
print(f"Movies : {ratings_filtered['movieId'].nunique():,}")
```

```python
tmdb  = pd.read_csv(TMDB_MOVIES)
creds = pd.read_csv(TMDB_CREDITS)

def parse_col(val, key='name', limit=None):
    try:
        items = ast.literal_eval(val)
        names = [i[key] for i in items if key in i]
        return names[:limit] if limit else names
    except:
        return []

tmdb['genres_list']   = tmdb['genres'].apply(parse_col)
tmdb['keywords_list'] = tmdb['keywords'].apply(parse_col)
creds['cast_list']    = creds['cast'].apply(lambda x: parse_col(x, limit=5))
creds['director']     = creds['crew'].apply(
    lambda x: next((i['name'] for i in ast.literal_eval(x)
                    if i.get('job')=='Director'), None)
    if pd.notnull(x) else None)

creds = creds.rename(columns={'movie_id':'id'})
tmdb_full = tmdb.merge(creds[['id','cast_list','director']], on='id', how='left')

def make_soup(row):
    def cn(n): return n.lower().replace(' ','') if n else ''
    genres   = ' '.join([cn(g) for g in (row['genres_list']   or [])])
    keywords = ' '.join([cn(k) for k in (row['keywords_list'] or [])])
    cast     = ' '.join([cn(c) for c in (row['cast_list']     or [])])
    director = cn(row['director']) * 3
    overview = str(row['overview']).lower() if pd.notnull(row['overview']) else ''
    return f"{genres} {keywords} {cast} {director} {overview}"

tmdb_full['soup'] = tmdb_full.apply(make_soup, axis=1)
print(f"tmdb_full shape: {tmdb_full.shape}")
print("Sample soup:", tmdb_full['soup'].iloc[0][:200])
```

```python
links  = pd.read_csv(LINKS_PATH)
ml_mov = pd.read_csv(MOVIES_ML)

links['tmdbId']  = pd.to_numeric(links['tmdbId'], errors='coerce')
links            = links.dropna(subset=['tmdbId'])
links['tmdbId']  = links['tmdbId'].astype(int)
tmdb_full['id']  = tmdb_full['id'].astype(int)

movies_df = ml_mov.merge(links[['movieId','tmdbId']], on='movieId', how='left')
movies_df = movies_df.merge(
    tmdb_full[['id','title','genres_list','keywords_list',
               'cast_list','director','overview','vote_average','vote_count','soup']],
    left_on='tmdbId', right_on='id', how='left'
).drop(columns=['id'], errors='ignore')
movies_df = movies_df.rename(columns={'title_x':'ml_title','title_y':'tmdb_title'})

from collections import Counter
all_genres = [g for gl in movies_df['genres_list'].dropna() for g in (gl if isinstance(gl,list) else [])]
gc = Counter(all_genres).most_common(12)
labels, vals = zip(*gc)
fig, axes = plt.subplots(1, 2, figsize=(12,4))
axes[0].barh(labels, vals, color='#5DCAA5'); axes[0].set_title('Top genres'); axes[0].invert_yaxis()
movies_df['vote_average'].dropna().hist(bins=20, ax=axes[1], color='#7F77DD', edgecolor='white')
axes[1].set_title('Vote average distribution')
plt.tight_layout(); plt.show()


ratings_filtered.to_csv(f'{OUTPUT_DIR}/ratings_filtered.csv', index=False)
movies_df.to_csv(f'{OUTPUT_DIR}/movies_df.csv', index=False)
tmdb_full.to_csv(f'{OUTPUT_DIR}/tmdb_full.csv', index=False)
print("Phase 2 complete. Files saved.")
print(f"  ratings_filtered : {len(ratings_filtered):,} rows")
print(f"  movies_df        : {len(movies_df):,} rows")
print(f"  tmdb_full        : {len(tmdb_full):,} rows")
```

```python
tmdb_clean = tmdb_full[tmdb_full['soup'].notna()].reset_index(drop=True)
tmdb_clean['soup'] = tmdb_clean['soup'].fillna('')

print("Building TF-IDF matrix")
tfidf = TfidfVectorizer(max_features=15000, ngram_range=(1,2),
                        min_df=2, stop_words='english')
tfidf_matrix = tfidf.fit_transform(tmdb_clean['soup'])
print(f"TF-IDF shape: {tfidf_matrix.shape}")

print("Computing cosine similarity")
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
print(f"Similarity matrix: {cosine_sim.shape}")

title_to_idx = pd.Series(tmdb_clean.index, index=tmdb_clean['title'].str.lower())

def get_content_recs(title, n=10):
    t = title.lower()
    if t not in title_to_idx: return f"'{title}' not found."
    idx  = title_to_idx[t]
    sims = sorted(enumerate(cosine_sim[idx]), key=lambda x:x[1], reverse=True)[1:n+1]
    res  = tmdb_clean[['title','genres_list']].iloc[[i for i,_ in sims]].copy()
    res['content_score'] = [round(s,4) for _,s in sims]
    return res

print(get_content_recs("The Dark Knight", n=5))
```

```python
def genre_precision_at_k(seed_title, k=10):
    t = seed_title.lower()
    if t not in title_to_idx: return None
    idx  = title_to_idx[t]
    seed_genres = set(tmdb_clean.iloc[idx]['genres_list'] or [])
    sims = sorted(enumerate(cosine_sim[idx]), key=lambda x:x[1], reverse=True)[1:k+1]
    hits = sum(1 for i,_ in sims
               if set(tmdb_clean.iloc[i]['genres_list'] or []) & seed_genres)
    return hits / k

test_movies = ["The Dark Knight","Toy Story","The Godfather","Inception","Pulp Fiction"]
scores = []
for m in test_movies:
    p = genre_precision_at_k(m)
    if p is not None:
        scores.append(p)
        print(f"  {m:<28} precision@10 = {p:.2f}")
print(f"\nMean precision@10 : {np.mean(scores):.2f}")
```

```python
from scipy.sparse import csr_matrix
from sklearn.decomposition import TruncatedSVD
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import mlflow, numpy as np

print("Building user-item matrix")
sample = ratings_filtered.sample(
    n=min(1_000_000, len(ratings_filtered)), random_state=42
).copy()

user_enc  = LabelEncoder()
movie_enc = LabelEncoder()
sample['user_idx']  = user_enc.fit_transform(sample['userId'])
sample['movie_idx'] = movie_enc.fit_transform(sample['movieId'])

n_users  = int(sample['user_idx'].max()) + 1
n_movies = int(sample['movie_idx'].max()) + 1
print(f"Matrix : {n_users:,} users × {n_movies:,} movies")

ui_matrix = csr_matrix(
    (sample['rating'].values.astype(np.float32),
     (sample['user_idx'].values, sample['movie_idx'].values)),
    shape=(n_users, n_movies)
)
print(f"Non-zero entries: {ui_matrix.nnz:,}")

try:
    mlflow.end_run()
except:
    pass

mlflow.set_experiment("cineiq-svd")
print("Fitting TruncatedSVD ")

with mlflow.start_run(run_name="sklearn-svd-baseline"):
    svd_model = TruncatedSVD(n_components=100, random_state=42)
    U  = svd_model.fit_transform(ui_matrix)  # (n_users, 100)
    Vt = svd_model.components_               # (100, n_movies)
    explained = float(svd_model.explained_variance_ratio_.sum())
    print(f"Explained variance : {explained:.4f}")
    mlflow.log_param("n_components", 100)
    mlflow.log_metric("explained_variance", round(explained, 4))

print(f"U  shape : {U.shape}")
print(f"Vt shape : {Vt.shape}")


print("\nReconstructing rating matrix")
reconstructed = U @ Vt  # (n_users, n_movies)
print(f"Reconstructed shape: {reconstructed.shape}")
def predict_rating(user_id, movie_id):
    if user_id not in user_enc.classes_:
        return 3.5
    if movie_id not in movie_enc.classes_:
        return 3.5
    u_idx = int(user_enc.transform([user_id])[0])
    m_idx = int(movie_enc.transform([movie_id])[0])
    pred  = reconstructed[u_idx, m_idx]
    return float(np.clip(pred, 0.5, 5.0))


print("\nEvaluating on 20% held-out test set")
_, test_s = train_test_split(sample, test_size=0.2, random_state=42)
test_s = test_s.copy()
test_s['pred'] = test_s.apply(
    lambda r: predict_rating(r['userId'], r['movieId']), axis=1
)

rmse = float(np.sqrt(mean_squared_error(test_s['rating'], test_s['pred'])))
mae  = float(np.abs(test_s['rating'] - test_s['pred']).mean())
print(f"SVD RMSE : {rmse:.4f}")
print(f"SVD MAE  : {mae:.4f}")

sample_pred = predict_rating(
    sample['userId'].iloc[0],
    sample['movieId'].iloc[0]
)
print(f"\nSanity check — userId={sample['userId'].iloc[0]}, "
      f"movieId={sample['movieId'].iloc[0]} → {sample_pred:.2f}")
```

```python
def normalize(series):
    mn, mx = series.min(), series.max()
    return (series - mn) / (mx - mn) if mx > mn else series * 0


def ensemble_recommend(seed_title, user_id, alpha=0.5, beta=0.5, n=10):
    t = seed_title.lower()
    if t not in title_to_idx:
        return f"'{seed_title}' not found."

    idx = title_to_idx[t]
    sims = sorted(enumerate(cosine_sim[idx]), key=lambda x: x[1], reverse=True)[1:100]

    cands = tmdb_clean.iloc[[i for i, _ in sims]].copy()
    cands['content_score'] = [s for _, s in sims]
    cands['tmdbId'] = pd.to_numeric(cands['id'], errors='coerce').astype('Int64')

    ml_ids = movies_df[['tmdbId', 'movieId']].dropna().copy()
    ml_ids['tmdbId'] = ml_ids['tmdbId'].astype('Int64')
    cands = cands.merge(ml_ids, on='tmdbId', how='left')

    def get_svd(mid):
        if pd.notnull(mid):
            return predict_rating(user_id, mid)
        return 3.5
    cands['svd_score'] = cands['movieId'].apply(get_svd)
    cands['cn'] = normalize(cands['content_score'])
    cands['sn'] = normalize(cands['svd_score'])
    cands['ensemble_score'] = alpha * cands['cn'] + beta * cands['sn']

    cols = ['title', 'genres_list', 'content_score', 'svd_score', 'ensemble_score']
    top = cands.nlargest(n, 'ensemble_score')[cols].reset_index(drop=True)
    top.index += 1
    return top


print(ensemble_recommend("The Dark Knight", user_id=1, n=5))
```

```python
import torch
import numpy as np
from transformers import (
    DistilBertTokenizer,
    DistilBertForSequenceClassification,
    TrainingArguments,
    Trainer
)
from torch.utils.data import Dataset as TorchDataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

OUTPUT_DIR = '/kaggle/working/cineiq'

print(f"GPU available : {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU name      : {torch.cuda.get_device_name(0)}")

MODEL_NAME = 'distilbert-base-uncased'
tokenizer  = DistilBertTokenizer.from_pretrained(MODEL_NAME)

tr_texts, val_texts, tr_labels, val_labels = train_test_split(
    reviews['clean_review'].tolist(),
    reviews['label'].tolist(),
    test_size=0.15,
    random_state=42,
    stratify=reviews['label']
)
print(f"Train samples : {len(tr_texts):,}")
print(f"Val samples   : {len(val_texts):,}")

class IMDBDataset(TorchDataset):
    def __init__(self, texts, labels, tok, max_len=256):
        self.enc = tok(
            texts,
            truncation=True,
            padding=True,
            max_length=max_len,
            return_tensors='pt'
        )
        self.labels = torch.tensor(labels)

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, i):
        item = {k: v[i] for k, v in self.enc.items()}
        item['labels'] = self.labels[i]
        return item

print("Tokenising datasets")
train_ds = IMDBDataset(tr_texts,  tr_labels,  tokenizer)
val_ds   = IMDBDataset(val_texts, val_labels, tokenizer)
print("Tokenisation complete.")

def compute_metrics(ep):
    preds = np.argmax(ep.predictions, axis=1)
    return {'accuracy': accuracy_score(ep.label_ids, preds)}

model = DistilBertForSequenceClassification.from_pretrained(
    MODEL_NAME, num_labels=2
)

args = TrainingArguments(
    output_dir=f'{OUTPUT_DIR}/bert-ckpt',
    num_train_epochs=3,
    per_device_train_batch_size=32,
    per_device_eval_batch_size=64,
    learning_rate=2e-5,
    weight_decay=0.01,
    eval_strategy='epoch',
    save_strategy='epoch',
    load_best_model_at_end=True,
    metric_for_best_model='accuracy',
    fp16=torch.cuda.is_available(),
    report_to='none',
    logging_steps=200
)
trainer = Trainer(
    model=model,
    args=args,
    train_dataset=train_ds,
    eval_dataset=val_ds,
    compute_metrics=compute_metrics
)

print("Training DistilBERT")
trainer.train()

eval_result = trainer.evaluate()
print(f"\nFinal eval accuracy : {eval_result['eval_accuracy']:.4f}")
print(f"Final eval loss     : {eval_result['eval_loss']:.4f}")

BERT_DIR = f'{OUTPUT_DIR}/distilbert-sentiment'
trainer.save_model(BERT_DIR)
tokenizer.save_pretrained(BERT_DIR)
print(f"\nModel saved → {BERT_DIR}")

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
model.eval()

def bert_sentiment(text, max_len=256):
    inp = tokenizer(
        text, return_tensors='pt',
        truncation=True, padding=True, max_length=max_len
    ).to(device)
    with torch.no_grad():
        logits = model(**inp).logits
    probs = torch.softmax(logits, dim=1).cpu().numpy()[0]
    score = float(probs[1] - probs[0])
    return {'score': round(score, 4),
            'label': 'positive' if score > 0 else 'negative'}

for ex in [
    "Absolutely brilliant. One of the best films ever made.",
    "Terrible movie. Complete waste of time.",
]:
    print(bert_sentiment(ex))
```

```python
BERT_DIR = f'{OUTPUT_DIR}/distilbert-sentiment'
trainer.save_model(BERT_DIR); tokenizer.save_pretrained(BERT_DIR)
print(f"Model saved → {BERT_DIR}")

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device); model.eval()

def bert_sentiment(text, max_len=256):
    inp = tokenizer(text, return_tensors='pt', truncation=True,
                    padding=True, max_length=max_len).to(device)
    with torch.no_grad():
        logits = model(**inp).logits
    probs = torch.softmax(logits, dim=1).cpu().numpy()[0]
    score = float(probs[1] - probs[0])
    return {'score': round(score,4),
            'label': 'positive' if score>0 else 'negative'}

def sentiment_rerank(recs_df, method='vader', weight=0.2):
    df = recs_df.copy()
    if method == 'vader':
        df['sent_score'] = df['title'].apply(
            lambda t: analyzer.polarity_scores(str(t))['compound'])
    else:
        df['sent_score'] = df['title'].apply(
            lambda t: bert_sentiment(str(t))['score'])
    df['sent_norm']  = (df['sent_score'] + 1) / 2
    df['final_score']= (1-weight)*df['ensemble_score'] + weight*df['sent_norm']
    return df.sort_values('final_score', ascending=False).reset_index(drop=True)

print("Sentiment re-ranker ready.")
for ex in ["Absolutely brilliant film!","Terrible waste of time."]:
    print(ex, "→", bert_sentiment(ex))
```

```python
def safe_parse(v):
    try: return ast.literal_eval(v) if isinstance(v,str) else (v or [])
    except: return []

tmdb_indexed = pd.read_csv(f'/kaggle/working/cineiq/tmdb_indexed.csv', index_col=0)
tmdb_indexed['genres_list'] = tmdb_indexed['genres_list'].apply(safe_parse)

def rule_explanation(seed_title, rec_title, content_score, svd_score=None):
    sm = tmdb_indexed[tmdb_indexed['title'].str.lower()==seed_title.lower()]
    rm = tmdb_indexed[tmdb_indexed['title'].str.lower()==rec_title.lower()]
    if sm.empty or rm.empty:
        return f"Recommended based on your interest in {seed_title}."
    sr, rr = sm.iloc[0], rm.iloc[0]
    sg = set(safe_parse(sr.get('genres_list',[])))
    rg = set(safe_parse(rr.get('genres_list',[])))
    sd = str(sr.get('director','') or '')
    rd = str(rr.get('director','') or '')
    parts = []
    if sd and rd and sd.lower()==rd.lower() and sd!='nan':
        parts.append(f"directed by the same filmmaker ({sd})")
    if sg & rg:
        parts.append("is also a " + " and ".join(list(sg&rg)[:2]) + " film")
    if content_score > 0.35:
        parts.append(f"has {content_score:.0%} content similarity")
    elif content_score > 0.15:
        parts.append(f"shares thematic elements with {seed_title}")
    if svd_score and svd_score >= 4.0:
        parts.append(f"predicted {svd_score:.1f}/5 for viewers like you")
    if not parts:
        parts.append("matches your viewing pattern")
    return "Recommended because it " + ", and ".join(parts[:3]) + "."

print(rule_explanation("The Dark Knight","Batman Begins", 0.72, 4.3))
```

```python
def full_pipeline(seed_title, user_id=1, alpha=0.5, beta=0.5,
                  sentiment_weight=0.15, n=10, use_lime=False):
    t = seed_title.lower()
    if t not in title_to_idx.index:
        print(f"'{seed_title}' not found."); return None

    idx  = title_to_idx[t]
    sims = sorted(enumerate(cosine_sim[idx]), key=lambda x:x[1], reverse=True)[1:80]

    cands = tmdb_clean.iloc[[i for i,_ in sims]].copy()
    cands['content_score'] = [s for _,s in sims]

    cands['tmdbId'] = pd.to_numeric(cands['id'], errors='coerce').astype('Int64')
    ml_ids = movies_df[['tmdbId','movieId']].dropna()
    ml_ids['tmdbId'] = ml_ids['tmdbId'].astype('Int64')
    cands = cands.merge(ml_ids, on='tmdbId', how='left')

    cands['svd_score'] = cands['movieId'].apply(
        lambda mid: predict_rating(user_id, mid) if pd.notnull(mid) else 3.5)

    cands['cn'] = normalize(cands['content_score'])
    cands['sn'] = normalize(cands['svd_score'])
    cands['ensemble_score'] = alpha*cands['cn'] + beta*cands['sn']

    cands['sent_raw']  = cands['title'].apply(
        lambda t2: analyzer.polarity_scores(str(t2))['compound'])
    cands['sent_norm'] = (cands['sent_raw'] + 1) / 2
    cands['final_score'] = ((1-sentiment_weight)*cands['ensemble_score']
                            + sentiment_weight*cands['sent_norm'])

    top = cands.nlargest(n, 'final_score').reset_index(drop=True)

    def make_exp(row):
        exp = rule_explanation(seed_title, row['title'],
                               row['content_score'], row['svd_score'])
        if use_lime and row['content_score'] < 0.2:
            _, lt = lime_explain(row['title'])
            if lt: exp = exp.rstrip('.') + '. ' + lt
        return exp

    top['explanation'] = top.apply(make_exp, axis=1)
    top.index += 1
    return top[['title','final_score','content_score','svd_score','sent_raw','explanation']]
for seed in ["The Dark Knight","Toy Story","The Godfather"]:
    print(f"\n{'='*55}\nSeed: {seed}")
    res = full_pipeline(seed, n=5)
    if res is not None:
        for _, row in res.iterrows():
            print(f"  {row['title']:<35} {row['final_score']:.3f}")
            print(f"  → {row['explanation']}")
```
